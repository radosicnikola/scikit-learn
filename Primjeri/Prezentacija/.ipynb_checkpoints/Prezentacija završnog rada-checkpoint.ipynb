### Priprema modula <a id="1"></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.offline as py
import plotly.graph_objs as go

py.init_notebook_mode(connected=True)
```

### Priprema podataka <a id="2"></a>

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

podaci_vino = datasets.load_wine()
podaci, oznake = podaci_vino.data, podaci_vino.target
podaci_tren, podaci_test, oznake_tren, oznake_test = train_test_split(podaci, oznake, test_size=0.2,
                                                                      stratify=oznake, random_state=1)
```

```python
df = pd.DataFrame(podaci, columns=podaci_vino.feature_names)
df.head(5)
```

### Pohlepno traženje hiper-parametara modela <a id="3"></a>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

pipe_lr = make_pipeline(StandardScaler(), PCA(n_components=3), LogisticRegression())

raspon = np.logspace(-3, 3, 500)
mogucnosti = {'logisticregression__penalty': ['l1', 'l2'], 'logisticregression__C': raspon}
gs = GridSearchCV(estimator=pipe_lr, param_grid=mogucnosti, scoring='accuracy',
                  cv=10, n_jobs=-1, verbose=True, return_train_score=True)
```

```python
gs.fit(podaci_tren, oznake_tren)
```

```python
np.round(gs.best_score_, 3)
```

```python
gs.best_params_
```

```python
est = gs.best_estimator_
```

```python
from sklearn.metrics import classification_report

print(classification_report(oznake_test, est.predict(podaci_test)))
```

```python
df = pd.DataFrame(est.predict_proba(podaci_test), columns=('klasa 0', 'klasa 1', 'klasa 2'))
df.head(5)
```

```python
print(est.predict(podaci_test)[:5])
```

### Važnosti glavnih komponenata <a id="4"></a>

```python
pca_all_components = PCA()
pca_all_components.fit(StandardScaler().fit_transform(podaci_tren), oznake_tren)

data = go.Bar(x=list(range(1,14)), y=pca_all_components.explained_variance_ratio_,
              opacity=0.8, name='Pojedinačno objašnjena varijanca')
cum_sum = go.Scatter(
    x=list(range(1,14)),
    y=np.cumsum(pca_all_components.explained_variance_ratio_),
    mode='lines+markers',
    name='Kulumativno objašnjena varijanca',
    line=dict(
        shape='hv'
    )
)
layout = go.Layout(
    title='',
    xaxis=dict(title='Indeks glavne komponente', tickmode='linear'),
    yaxis=dict(title='Objašnjeni omjer varijance'),
    showlegend=True
)
fig = go.Figure(data=[data, cum_sum], layout=layout)
py.iplot(fig, filename='basic-bar')
```

### Vizualizacija uzoraka i granica odlučivanja <a id="5"></a>

```python
boje = ['yellow', 'blue', 'red']
def prikaz3d(podaci_tren, podaci_test, oznake_tren, oznake_test, labele):
    data = []
    for klasa, oznaka, boja in zip(range(3), np.unique(oznake_tren), boje):
        trace1 = go.Scatter3d(
            x=podaci_tren[oznake_tren==oznaka, 0],
            y=podaci_tren[oznake_tren==oznaka, 1],
            z=podaci_tren[oznake_tren==oznaka, 2],
            mode='markers',
            marker=dict(
                size=8,
                line=dict(
                    width=0.5
                ),
                color=boja,
                opacity=0.8
            ),
            name=f'klasa {klasa} tren'
        )
        trace2 = go.Scatter3d(
            x=podaci_test[oznake_test==oznaka, 0],
            y=podaci_test[oznake_test==oznaka, 1],
            z=podaci_test[oznake_test==oznaka, 2],
            mode='markers',
            marker=dict(
                size=8,
                line=dict(
                    width=0.5
                ),
                color=boja,
                opacity=0.8,
                symbol='x'
            ),
            name=f'klasa {klasa} test'
        )
        data.append(trace1)
        data.append(trace2)
    layout = go.Layout(
        margin=dict(
            l=0,
            r=0,
            b=0,
            t=0
        ),
        title='Oznake',
        scene=dict(
            xaxis=dict(title=labele[0]),
            yaxis=dict(title=labele[1]),
            zaxis=dict(title=labele[2])
        )
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig, filename='simple-3d-scatter')
prikaz3d(podaci_tren, podaci_test, oznake_tren, oznake_test, ('PC1', 'PC2', 'PC3'))
```

```python
def izracun_granica(podaci, klasifikator, rez):

    x1_min, x1_max = podaci[:, 0].min() - 1, podaci[:, 0].max() + 1
    x2_min, x2_max = podaci[:, 1].min() - 1, podaci[:, 1].max() + 1
    x3_min, x3_max = podaci[:, 2].min() - 1, podaci[:, 2].max() + 1
    
    xx1, xx2, xx3 = np.meshgrid(np.arange(x1_min, x1_max, rez),
                                np.arange(x2_min, x2_max, rez),
                                np.arange(x3_min, x3_max, rez))
    pom = np.array([xx1.ravel(), xx2.ravel(), xx3.ravel()]).T
    boje = klasifikator.predict(pom)
    boje = boje.reshape(xx1.shape)
    return xx1, xx2, xx3, boje

podaci_tren_obrada = PCA(n_components=3).fit_transform(StandardScaler().fit_transform(podaci_tren))
lr = LogisticRegression(penalty='l2', C=0.1).fit(podaci_tren_obrada, oznake_tren)
xx1, xx2, xx3, boje = izracun_granica(podaci_tren_obrada, lr, rez=0.5)

def prikaz_granica3d(xx1, xx2, xx3, boje, b, labele):
    data = []
    for i in range(0, len(np.unique(boje))):
        trace1 = go.Scatter3d(
            x=xx1[boje==i],
            y=xx2[boje==i],
            z=xx3[boje==i],
            mode='markers',
            marker=dict(
                size=7,
                line=dict(
                    width=0.5
                ),
                color=b[i],
                opacity=0.6
            ),
            name=f'klasa {i}'
        )
        data.append(trace1)
    layout = go.Layout(
        margin=dict(
            l=0,
            r=0,
            b=0,
            t=0
        ),
        title='Granice klasnih pripadnosti uzoraka',
        scene=dict(
            xaxis=dict(title=labele[0]),
            yaxis=dict(title=labele[1]),
            zaxis=dict(title=labele[2])
        )
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig, filename='simple-3d-scatter')

b = {0:'yellow', 1:'blue', 2:'red'}
labele = ('PC1', 'PC2', 'PC3')
prikaz_granica3d(xx1, xx2, xx3, boje, b, labele)
```

```python
from mpl_toolkits.mplot3d import Axes3D
def prikaz_granica3d_1(xx1, xx2, xx3, boje, b, labele):
    fig = plt.figure(figsize=(10,10))
    ax = Axes3D(fig)
    for i in range(0, len(np.unique(boje))):
        ax.scatter(xx1[boje==i], xx2[boje==i], xx3[boje==i], c=b[i], marker='o', s=3, label=i, alpha=0.1)
    ax.set_xlabel(labele[0])
    ax.set_ylabel(labele[1])
    ax.set_zlabel(labele[2])
    plt.legend(loc="upper right", scatterpoints=7, markerscale=10.0, fontsize=10, prop={'size': 16})
    ax.view_init(25, 60)
    plt.show()

b = {0:'yellow', 1:'blue', 2:'red'}
labele = ('PC1', 'PC2', 'PC3')
xx1, xx2, xx3, boje = izracun_granica(podaci_tren_obrada, lr, rez=0.1)
prikaz_granica3d_1(xx1, xx2, xx3, boje, b, labele)
```

### Krivulje učenja i validacije <a id="6"></a>

```python
from sklearn.model_selection import learning_curve

velicine_skupa, rez_tren, rez_test = learning_curve(estimator=est, X=podaci_tren, y=oznake_tren,
                                                        train_sizes=np.linspace(0.1, 1.0, 20), cv=10, n_jobs=-1)
```

```python
def learn_curve(velicine_skupa, rez_tren, rez_test):
    tren_sred = np.mean(rez_tren, axis=1)
    tren_sdev = np.std(rez_tren, axis=1)
    test_sred = np.mean(rez_test, axis=1)
    test_sdev = np.std(rez_test, axis=1)
    
    upper_bound = go.Scatter(
        name='std + dev',
        x=velicine_skupa,
        y=tren_sred + tren_sdev,
        mode='lines',
        marker=dict(color="red"),
        line=dict(width=0),
        fillcolor='rgba(255, 0, 0, 0.15)',
        fill='tonexty',
        showlegend=False)

    trace = go.Scatter(
        name='Srednje vrijednosti <br>preciznosti treniranja modela',
        x=velicine_skupa,
        y=tren_sred,
        mode='lines+markers',
        line=dict(color='red'),
        fillcolor='rgba(255, 0, 0, 0.15)',
        fill='tonexty')

    lower_bound = go.Scatter(
        name='std - dev',
        x=velicine_skupa,
        y=tren_sred - tren_sdev,
        marker=dict(color="red"),
        line=dict(width=0),
        mode='lines',
        showlegend=False)
    
    upper_bound1 = go.Scatter(
        name='std + dev',
        x=velicine_skupa,
        y=test_sred + test_sdev,
        mode='lines',
        marker=dict(color="blue"),
        line=dict(width=0),
        fillcolor='rgba(0, 0, 255, 0.15)',
        fill='tonexty',
        showlegend=False)

    trace1 = go.Scatter(
        name='Srednje vrijednosti <br>preciznosti validacija modela',
        x=velicine_skupa,
        y=test_sred,
        mode='lines+markers',
        line=dict(color='blue'),
        fillcolor='rgba(0, 0, 255, 0.15)',
        fill='tonexty')

    lower_bound1 = go.Scatter(
        name='std - dev',
        x=velicine_skupa,
        y=test_sred - test_sdev,
        marker=dict(color="blue"),
        line=dict(width=0),
        mode='lines',
        showlegend=False)
    data = [lower_bound, trace, upper_bound, lower_bound1, trace1, upper_bound1]
    layout = go.Layout(
        xaxis=dict(title='Broj uzoraka za treniranje'),
        yaxis=dict(title='Preciznost', range=[0.8, 1.05]),
        title='Krivulje učenja'
    )

    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig, filename='pandas-continuous-error-bars')
```

```python
learn_curve(velicine_skupa, rez_tren, rez_test)
```

```python
from sklearn.model_selection import validation_curve

raspon = np.logspace(-3, 3, 500)
rez_tren, rez_test = validation_curve(estimator=est, X=podaci_tren, y=oznake_tren, cv=10, n_jobs=-1,
                                                      param_name='logisticregression__C', param_range=raspon)
```

```python
def valid_curve(raspon, rez_tren, rez_test):
    tren_sred = np.mean(rez_tren, axis=1)
    tren_sdev = np.std(rez_tren, axis=1)
    test_sred = np.mean(rez_test, axis=1)
    test_sdev = np.std(rez_test, axis=1)
    
    upper_bound = go.Scatter(
        name='std + dev',
        x=raspon,
        y=tren_sred + tren_sdev,
        mode='lines',
        marker=dict(color="red"),
        line=dict(width=0),
        fillcolor='rgba(255, 0, 0, 0.15)',
        fill='tonexty',
        showlegend=False)

    trace = go.Scatter(
        name='Srednje vrijednosti <br>preciznosti treniranja modela',
        x=raspon,
        y=tren_sred,
        mode='lines',
        line=dict(color='red'),
        fillcolor='rgba(255, 0, 0, 0.15)',
        fill='tonexty')

    lower_bound = go.Scatter(
        name='std - dev',
        x=raspon,
        y=tren_sred - tren_sdev,
        marker=dict(color="red"),
        line=dict(width=0),
        mode='lines',
        showlegend=False)
    
    upper_bound1 = go.Scatter(
        name='std + dev',
        x=raspon,
        y=test_sred + test_sdev,
        mode='lines',
        marker=dict(color="blue"),
        line=dict(width=0),
        fillcolor='rgba(0, 0, 255, 0.15)',
        fill='tonexty',
        showlegend=False)

    trace1 = go.Scatter(
        name='Srednje vrijednosti <br>preciznosti validacija modela',
        x=raspon,
        y=test_sred,
        mode='lines',
        line=dict(color='blue'),
        fillcolor='rgba(0, 0, 255, 0.15)',
        fill='tonexty')

    lower_bound1 = go.Scatter(
        name='std - dev',
        x=raspon,
        y=test_sred - test_sdev,
        marker=dict(color="blue"),
        line=dict(width=0),
        mode='lines',
        showlegend=False)
    data = [lower_bound, trace, upper_bound, lower_bound1, trace1, upper_bound1]
    layout = go.Layout(
        xaxis=dict(title='Vrijednost hiperparametra <i>C</i>', type='log'),
        yaxis=dict(title='Preciznost', range=[0.8, 1.05]),
        title='Krivulje validacije'
    )

    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig, filename='pandas-continuous-error-bars')
```

```python
valid_curve(raspon, rez_tren, rez_test)
```
